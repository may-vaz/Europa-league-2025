```python
top_s = man_pl[man_pl['G']>0].sort_values(by='G',ascending=False).reset_index(drop=True)
print(top_s[['Name','G']])
```

```python

top_s['acc%'] = (top_s['ST']/top_s['SH'])*100
top_s[['Name','SH','ST','acc%']].sort_values(by='acc%',ascending=False)
```

```python
man_gk.head(2)
```

```python
man_gk['s_p'] = (man_gk['Saves']/(man_gk['Saves']+man_gk['Goals Against']))*100 #save percentage
man_gk['con'] = man_gk['Goals Against']/man_gk['Appearances'] #goals conceded per match
man_gk['saved']= man_gk['Saves']/man_gk['Appearances'] #goals saved per match


man_gk[['Name','s_p','con','saved']].round(2).head(2)
```

TOTTENHAM


```python
top_st = tot_pl[tot_pl['G']>0].sort_values(by='G',ascending=False).reset_index(drop=True)
print(top_st[['Name','G']])
```

```python
top_st['acc%'] = (top_st['ST']/top_st['SH'])*100
top_st[['Name','SH','ST','acc%']].sort_values(by='acc%',ascending=False)
```

```python
tot_gk.head()
```

```python
tot_gk['s_p'] = (tot_gk['SV']/(tot_gk['SV']+tot_gk['GA']))*100 #save percentage
tot_gk['con'] = tot_gk['GA']/tot_gk['APP'] #goals conceded per match
tot_gk['saved']= tot_gk['SV']/tot_gk['APP'] #goals saved per match


tot_gk[['Name','APP','s_p','con','saved']].round(2)
```
